```python
def setup_driver():
    """
        Function: 
            Setup the Google Chrome browser driver configurations for printing to PDF.
        Outputs:
            driver: the configured driver
    """

    appState = {
        "recentDestinations": [
            {
                "id": "Save as PDF",
                "origin": "local",
                "account": ""
            }
        ],
        "selectedDestinationId": "Save as PDF",
        "version": 2,
        "isHeaderFooterEnabled":False
    }
    
    chrome_path = "C:/Program Files/Google-OLD/chrome-win64/chrome.exe"
    save_to_path = "C:/Users/RDebe/Documents/0MSC DSBE/0Thesis/Prototypes/data"


    chrome_options = Options()
    chrome_options.binary_location = chrome_path  

    profile = {'printing.print_preview_sticky_settings.appState': json.dumps(appState),
                'savefile.default_directory': save_to_path}

    #chrome_options.add_argument('--headless') --> disable because of print preview
    #chrome_options.add_argument('--disable-gpu')

    # set language to english
    chrome_options.add_argument('--lang=en-US')  

    # set download directory and enable download prompt
    chrome_options.add_experimental_option('prefs', profile)

    chrome_options.add_argument('--kiosk-printing')


    driver = webdriver.Chrome(options=chrome_options)
   
    return driver
```

```python
def save_as_pdf(driver, url, index):
    """
        function to save the current page as a pdf
        gets url and index as input
        window prints the page 
    """


    try:
        driver.get(url)
        time.sleep(0.5)  
        driver.execute_script('window.print();')
        time.sleep(0.5)
        print(f"Saved: {url}")
    except Exception as e:
        print(f"Failed to save {url}: {str(e)}")
```

```python
def main(urls):
    """
    Loop through the urls and save them as pdf
    """
    driver = setup_driver()
    for index, url in enumerate(urls):
        save_as_pdf(driver, url, index)
    driver.quit()
```
